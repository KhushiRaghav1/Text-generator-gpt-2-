```python
from transformers import pipeline
```

#Load model


```python
generator = pipeline("text-generation",model="gpt2")
```

#give input prompt


```python
prompt = "story of a crow on a sunny day in a town "
```

#generate text

```python
result=generator(prompt,max_length=50)
```

#print output


```python
print(result[0]['generated_text'])
```
